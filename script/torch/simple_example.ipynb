```
dim = 2
sigma = 2.
sigma1 = 2.
nb_pts = 2

gd_trans = torch.tensor([[-2., 0.], [2., 0.]], requires_grad=True).view(-1)
mom_trans = torch.tensor([[0., -0.5], [0., 0.5]], requires_grad=True).view(-1)

gd_trans1 = torch.tensor([[-2., 0.], [2., 0.]], requires_grad=True).view(-1)
mom_trans1 = torch.tensor([[0., -0.5], [0., 0.5]], requires_grad=True).view(-1)

trans = dm.deformationmodules.Translations(dm.manifold.Landmarks(dim, nb_pts, gd=gd_trans, cotan=mom_trans), sigma)
trans1 = dm.deformationmodules.Translations(dm.manifold.Landmarks(dim, nb_pts, gd=gd_trans1, cotan=mom_trans1), sigma1)
hamiltonian = dm.hamiltonian.Hamiltonian([trans, trans1])
```
